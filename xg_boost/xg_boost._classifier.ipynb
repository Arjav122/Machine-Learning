```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score ,classification_report , confusion_matrix
import xgboost as xgb
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url ,sep=';')
```

```python
df.head()
```

```python
df['quality'] = df['quality'].apply(lambda x : 1 if x>=7 else 0 )
```

```python
df.head()
```

```python
df.describe()
```

```python
x = df.drop("quality" , axis=1)
y = df['quality']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2 , random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
xgb_classifier = xgb.XGBClassifier(objective="binary:logistic",eval_metric = "logloss")
```

```python
xgb_classifier.fit(x_train_scaled,y_train)
```

```python
xgb_classifier.predict(x_test_scaled)
```

```python
accuracy_score(y_test,xgb_classifier.predict(x_test_scaled))
```

```python
confusion_matrix(y_test,xgb_classifier.predict(x_test_scaled))
```

```python
print(classification_report(y_test,xgb_classifier.predict(x_test_scaled)))
```

```python
param_grid = {
    "n_estimators": [100, 200, 300],
    "learning_rate": [0.01, 0.05, 0.1],
    "max_depth": [3, 5, 7],
    "min_child_weight": [1, 3, 5],
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.7, 0.8, 1.0],
    "gamma": [0, 0.1, 0.2]
}
```

```python
grid_search = GridSearchCV(xgb.XGBClassifier(objective="binary:logistic",eval_metric = "logloss"),param_grid,cv =3 , n_jobs=6,verbose=2)
```

```python
grid_search.fit(x_train_scaled,y_train)
```

```python
grid_search.best_params_
```

```python
xgb_new = xgb.XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    min_child_weight=1,
    gamma=0.1,
    subsample=1.0,
    colsample_bytree=1.0,
    random_state=42
)
```

```python
xgb_new.fit(x_train_scaled, y_train)
```

```python
xgb_new.predict(x_test_scaled)
```

```python
accuracy_score(y_test, xgb_new.predict(x_test_scaled))
```
